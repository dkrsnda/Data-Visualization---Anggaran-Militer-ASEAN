```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('Tabel-Nabila-Pivot.csv', sep=";")
df2 = pd.read_csv('Tabel-Nabila-Pivot2.csv', sep=";")
```

```python
plt.title('Anggaran Militer 1951-2020')
plt.xlabel('Tahun')
plt.ylabel('Anggaran dalam Jutaan USD')

plt.plot(df1.Year, df1.USA, label='Amerika Serikat')
plt.plot(df1.Year, df1.Brunai)
plt.plot(df1.Year, df1.Filipina)
plt.plot(df1.Year, df1.Indonesia)
plt.plot(df1.Year, df1.Kamboja)
plt.plot(df1.Year, df1.Laos)
plt.plot(df1.Year, df1.Malaysia)
plt.plot(df1.Year, df1.Myanmar)
plt.plot(df1.Year, df1.Singapura)
plt.plot(df1.Year, df1.Thailand)
plt.plot(df1.Year, df1.Vietnam)
plt.plot(df1.Year, df1.China, label='China')
plt.plot(df1.Year, df1.Australia)

plt.legend()
```
