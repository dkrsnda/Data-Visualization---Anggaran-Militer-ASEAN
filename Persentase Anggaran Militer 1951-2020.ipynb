```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('Tabel-Nabila-Pivot2.csv', sep=";")
```

```python
plt.figure(figsize=(20,7))


plt.title('Persentase Anggaran Militer dari Total GDP (1951-2020)')
plt.xlabel('Tahun')
plt.ylabel('%')

plt.plot(df1.Year, df1.USA, label='Amerika Serikat')
plt.plot(df1.Year, df1.China, label='China')
plt.plot(df1.Year, df1.Brunei, label='Brunei Darussalam')
plt.plot(df1.Year, df1.Filipina, label='Filipina')
plt.plot(df1.Year, df1.Indonesia, label='Indonesia')
plt.plot(df1.Year, df1.Kamboja, label='Kamboja')
plt.plot(df1.Year, df1.Laos, label='Laos')
plt.plot(df1.Year, df1.Malaysia, label='Malaysia')
plt.plot(df1.Year, df1.Myanmar, label='Myanmar')
plt.plot(df1.Year, df1.Singapura, label='Singapura')
plt.plot(df1.Year, df1.Thailand, label='Thailand')
plt.plot(df1.Year, df1.Vietnam, label='Vietnam')
plt.plot(df1.Year, df1.Australia, label='Australia')

plt.legend()
```
